```python
import math

class Cell:
    def __init__(self, row, col):
        self.row = row
        self.col = col
        self.name = f'R{row}C{col}'
        self.neighbors = [None] * 4  # [up, right, down, left]
        self.connections = [False] * 4  # [up, right, down, left]

class Grid:
    def __init__(self, m, n):
        self.rows = m
        self.cols = n
        self.cells = [[Cell(i, j) for j in range(n)] for i in range(m)]
        for i in range(m):
            for j in range(n):
                if i > 0:
                    self.cells[i][j].neighbors[0] = self.cells[i-1][j]  # up
                if j < n-1:
                    self.cells[i][j].neighbors[1] = self.cells[i][j+1]  # right
                if i < m-1:
                    self.cells[i][j].neighbors[2] = self.cells[i+1][j]  # down
                if j > 0:
                    self.cells[i][j].neighbors[3] = self.cells[i][j-1]  # left

    def reset_connections(self):
        for i in range(self.rows):
            for j in range(self.cols):
                self.cells[i][j].connections = [False] * 4

    def apply_walls(self, boundary_statuses, num_walls):
        self.reset_connections()
        m, n = self.rows, self.cols
        wall_idx = 0
        for i in range(m):
            for j in range(2 * n - 1):
                if wall_idx >= num_walls:
                    break
                if i < m - 1 or j % 2 != 0:
                    bit = boundary_statuses[wall_idx]
                    wall_idx += 1
                    cell = self.cells[i][j // 2]
                    if i < m - 1 and j % 2 == 0:  # Horizontal wall
                        if bit:
                            cell.connections[2] = True
                            cell.neighbors[2].connections[0] = True
                    elif j % 2 != 0:  # Vertical wall
                        if bit:
                            cell.connections[1] = True
                            cell.neighbors[1].connections[3] = True

def connections(grid, x, y, visited):
    cell = grid.cells[x][y]
    if cell.name in visited:
        return
    visited.add(cell.name)
    for i, neighbor in enumerate(cell.neighbors):
        if neighbor and cell.connections[i]:
            connections(grid, neighbor.row, neighbor.col, visited)

def connected_regions(grid):
    regions = []
    visited = set()
    for i in range(grid.rows):
        for j in range(grid.cols):
            if grid.cells[i][j].name not in visited:
                region = set()
                connections(grid, i, j, region)
                regions.append(region)
                visited.update(region)
    return regions

def calc_score(regions, m, n):
    score = sum(1 for region in regions if len(region) <= 3)
    return score / (m * n)

def calc_exp(m, n, p):
    if m == 1 and n == 1:
        return 1.0
    num_walls = 2 * m * n - m - n
    grid = Grid(m, n)
    total_exp = 0.0
    num_configs = 1 << num_walls
    for i in range(num_configs):
        boundary_statuses = [(i >> j) & 1 for j in range(num_walls)]
        num_removed = sum(boundary_statuses)
        grid.apply_walls(boundary_statuses, num_walls)
        regions = connected_regions(grid)
        score = calc_score(regions, m, n)
        prob = (p ** num_removed) * ((1 - p) ** (num_walls - num_removed))
        total_exp += score * prob
    return total_exp

# Input processing
q = int(input().strip())
for _ in range(q):
    m, n, p = map(float, input().split())
    m, n = int(m), int(n)
    result = calc_exp(m, n, p)
    print(f"{result:.10f}")
```
