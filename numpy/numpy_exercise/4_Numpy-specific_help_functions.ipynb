Q3 Get help information for numpy many diffrent function.

```python
# prompt: Get help information for numpy many diferent function.

# Getting help for multiple functions
# You can list the functions you are interested in and then loop through them
functions_to_help = [np.dot, np.linalg.eig, np.mean, np.sum]

for func in functions_to_help:
  print(f"Help information for {func.__name__}:")
  help(func)
  print("-" * 30)
```
