```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car object
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat object
plane1 = Plane("Boeing", "747")     #Create a Plane object

for x in (car1, boat1, plane1):
  x.move()
```

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Move!")

class Car(Vehicle):
  pass

class Boat(Vehicle):
  def move(self):
    print("Sail!")

class Plane(Vehicle):
  def move(self):
    print("Fly!")

car1 = Car("Ford", "Mustang")       #Create a Car object
boat1 = Boat("Ibiza", "Touring 20") #Create a Boat object
plane1 = Plane("Boeing", "747")     #Create a Plane object

for x in (car1, boat1, plane1):
  print(x.brand)
  print(x.model)
  x.move()
```

```python
#1. Simple Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def bark(self):
        print("Woof!")

my_dog = Dog("Fido")
print(my_dog.name)  # Output: Fido
my_dog.bark()  # Output: Woof!

#2. Multiple Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

class Mammal:
    def eat(self):
        print("Eating...")

class Dog(Animal, Mammal):
    def bark(self):
        print("Woof!")

my_dog = Dog("Fido")
print(my_dog.name)  # Output: Fido
my_dog.eat()  # Output: Eating...
my_dog.bark()  # Output: Woof!

#3. Multilevel Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

class Mammal(Animal):
    def eat(self):
        print("Eating...")

class Dog(Mammal):
    def bark(self):
        print("Woof!")

my_dog = Dog("Fido")
print(my_dog.name)  # Output: Fido
my_dog.eat()  # Output: Eating...
my_dog.bark()  # Output: Woof!

#4. Hierarchical Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def bark(self):
        print("Woof!")

class Cat(Animal):
    def meow(self):
        print("Meow!")

my_dog = Dog("Fido")
my_cat = Cat("Whiskers")
print(my_dog.name)  # Output: Fido
my_dog.bark()  # Output: Woof!
print(my_cat.name)  # Output: Whiskers
my_cat.meow()  # Output: Meow!

#5. Inheritance with Constructors

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

my_dog = Dog("Fido", "Golden Retriever")
print(my_dog.name)  # Output: Fido
print(my_dog.breed)  # Output: Golden Retriever

#6. Inheritance with Methods Overriding

class Animal:
    def sound(self):
        print(" Generic animal sound")

class Dog(Animal):
    def sound(self):
        print("Woof!")

my_dog = Dog()
my_dog.sound()  # Output: Woof!

#7. Inheritance with Abstract Methods

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Woof!")

my_dog = Dog()
my_dog.sound()  # Output: Woof!
#8. Inheritance with Private Members

class Animal:
    def __init__(self, name):
        self.__name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.__breed = breed

    def get_name(self):
        return self._Animal__name

    def get_breed(self):
        return self.__breed

my_dog = Dog("Fido", "Golden Retriever")
print(my_dog.get_name())  # Output: Fido
print(my_dog.get_breed())  # Output: Golden Retriever

#9. Inheritance with Protected Members

class Animal:
    def __init__(self, name):
        self._name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self._breed = breed

    def get_name(self):
        return self._name

    def get_breed(self):
        return self._breed

my_dog = Dog("Fido", "Golden Retriever")
print(my_dog.get_name())  # Output: Fido
print(my_dog.get_breed())  # Output: Golden Retriever

#10. Inheritance with Static Methods

class Animal:
    @staticmethod
    def get_species():
        return "Animal"







```
